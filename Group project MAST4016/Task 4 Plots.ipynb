# January 2021 Trend in Recoveries

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load data for recoveries
recoveries = pd.read_csv('time_series_covid19_recovered_global.csv')

# Filter data for the Netherlands without provinces
netherlands_recoveries = recoveries[(recoveries['Country/Region'] == 'Netherlands') & recoveries['Province/State'].isna()]

# Drop unnecessary columns
netherlands_recoveries = netherlands_recoveries.drop(columns=['Lat', 'Long', 'Province/State'])

# Transpose the dataframe so that dates become rows and provinces/countries become columns
netherlands_recoveries = netherlands_recoveries.melt(id_vars='Country/Region', var_name='Date', value_name='Recoveries')

# Convert 'Date' column to datetime type
netherlands_recoveries['Date'] = pd.to_datetime(netherlands_recoveries['Date'])

# Extract month and year
netherlands_recoveries['Month'] = netherlands_recoveries['Date'].dt.month
netherlands_recoveries['Year'] = netherlands_recoveries['Date'].dt.year

# Filter data for January of each year (2020, 2021, 2022, 2023)
january_recoveries_2021 = netherlands_recoveries[(netherlands_recoveries['Month'] == 1) & (netherlands_recoveries['Year'] == 2021)]

# Find the minimum number of recoveries on the first day of January for each year
start_recoveries_2021 = january_recoveries_2021.groupby('Year')['Recoveries'].min().iloc[0]

# Group data by date and calculate total recoveries for each January
january_recoveries_total_2021 = january_recoveries_2021.groupby('Date')['Recoveries'].sum()

# Plotting
plt.figure(figsize=(14, 10))

# Plot January 2021 recoveries trend
plt.subplot(2, 2, 2)
plt.plot(january_recoveries_total_2021.index, january_recoveries_total_2021.values, color='green')
plt.title('January 2021 Trend in Recoveries')
plt.xlabel('Date')
plt.ylabel('Recoveries')
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability
plt.ylim(start_recoveries_2021, january_recoveries_total_2021.max() * 1.1)  # Set y-axis limits

plt.tight_layout()
plt.show()
```

# Trend of Deaths between Winter 2020 and the winter 2021

```python

# Load data for deaths
deaths = pd.read_csv('time_series_covid19_deaths_global.csv')

# Filter data for the Netherlands without provinces
netherlands_deaths = deaths[(deaths['Country/Region'] == 'Netherlands') & deaths['Province/State'].isna()]

# Drop unnecessary columns
netherlands_deaths = netherlands_deaths.drop(columns=['Lat', 'Long', 'Province/State'])

# Transpose the dataframe so that dates become rows and provinces/countries become columns
netherlands_deaths = netherlands_deaths.melt(id_vars='Country/Region', var_name='Date', value_name='Deaths')

# Convert 'Date' column to datetime type
netherlands_deaths['Date'] = pd.to_datetime(netherlands_deaths['Date'])

# Filter data for the year 2021
deaths_2021 = netherlands_deaths[(netherlands_deaths['Date'] >= '2020-12-22') & (netherlands_deaths['Date'] <= '2021-12-22')]

# Group data by date and calculate total deaths for each day in 2021
deaths_total_2021 = deaths_2021.groupby('Date')['Deaths'].sum()

# Plotting
plt.figure(figsize=(10, 6))

# Plot deaths trend for 2021
plt.plot(deaths_total_2021.index, deaths_total_2021.values, color='red', label='Deaths')

plt.title('Trend of Deaths between Winter 2020 and the winter 2021')
plt.xlabel('Date')
plt.ylabel('Number of Deaths')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()

```

# Trend of Recoveries between WINTER 2020 AND WINTER 2021


```python


# Load data for recoveries
recoveries = pd.read_csv('time_series_covid19_recovered_global.csv')

# Filter data for the Netherlands without provinces
netherlands_recoveries = recoveries[(recoveries['Country/Region'] == 'Netherlands') & recoveries['Province/State'].isna()]

# Drop unnecessary columns
netherlands_recoveries = netherlands_recoveries.drop(columns=['Lat', 'Long', 'Province/State'])

# Transpose the dataframe so that dates become rows and provinces/countries become columns
netherlands_recoveries = netherlands_recoveries.melt(id_vars='Country/Region', var_name='Date', value_name='Recoveries')

# Convert 'Date' column to datetime type
netherlands_recoveries['Date'] = pd.to_datetime(netherlands_recoveries['Date'])

# Filter data for the period between winter 2020 and winter 2021
winter_2020_start = '2020-12-22'
winter_2021_end = '2021-12-20'
recoveries_winter_to_winter = netherlands_recoveries[(netherlands_recoveries['Date'] >= winter_2020_start) & (netherlands_recoveries['Date'] <= winter_2021_end)]

# Group data by date and calculate total recoveries for each day
recoveries_total = recoveries_winter_to_winter.groupby('Date')['Recoveries'].sum()

# Plotting
plt.figure(figsize=(10, 6))

# Plot recoveries trend for the period between winter 2020 and winter 2021
plt.plot(recoveries_total.index, recoveries_total.values, color='green', label='Recoveries')

plt.title('Trend of Recoveries between Winter 2020 and Winter 2021')
plt.xlabel('Date')
plt.ylabel('Number of Recoveries')
plt.legend()
plt.xticks(rotation=45)  # Rotate x-axis labels for better readability

plt.tight_layout()
plt.show()
```
